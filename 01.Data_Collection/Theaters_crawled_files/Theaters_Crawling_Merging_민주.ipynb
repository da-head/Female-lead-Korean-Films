# 상영내역 데이터 + 상영관정보 데이터 merge
'상영관정보'에서 관별 좌석 수 데이터를 가져와서 '상영내역'에 붙인 후 저장.  
저장된 csv 파일은 'Theaters_merged_files' 폴더에 있음.

```python
code[150:300]
# ['영화상영관명_new'] = code['영화상영관명'].apply(lambda x: x.replace(' ', '_'))
```

```python
# for nm in code['영화상영관명'][150:300]:
for nm in code['영화상영관명']:
    # 안 받아진 파일    160(완료), 192, 215, 233: 235, 290
    print(nm)
# 위에서 지정해준 인덱스 숫자 n과 m을 사용해 code['영화상영관명'][n:m] 으로 변경

    theaNm = nm.strip().replace(' ', '%20').replace('(', '%28').replace(')', '%29')
    theaNm_detail = nm.strip().replace(' ', '_20').replace('(', '_28').replace(')', '_29')
    time = pd.read_html(f'상영내역-{theaNm}.html')[1]
    capacity = pd.read_html(f'상영관정보/상영관정보-{theaNm_detail}.html', encoding='utf-8')[1].T.drop('스크린코드', axis=0).rename(columns={0:'상영관', 1:'좌석수'})
    time.merge(capacity, how='left', on='상영관').to_csv(f'../Theaters_merged_files/{theaNm}.csv', index=False)
```

```python
# 크롤링 에러났던 상영관 데이터 (csv파일로 저장한 것 불러와서) 다시 merge

for nm in code['영화상영관명'][-2:]:
    print(nm)

    theaNm = nm.strip().replace(' ', '%20').replace('(', '%28').replace(')', '%29')
    theaNm_detail = nm.strip().replace(' ', '_20').replace('(', '_28').replace(')', '_29')
    time = pd.read_csv(f'상영내역-{theaNm}.csv')
    capacity = pd.read_html(f'상영관정보/상영관정보-{theaNm_detail}.html', encoding='utf-8')[1].T.drop('스크린코드', axis=0).rename(columns={0:'상영관', 1:'좌석수'})
    time.merge(capacity, how='left', on='상영관').to_csv(f'../Theaters_merged_files/{theaNm}.csv', index=False)
```

# Merge

```python
screen=pd.read_csv('screen_updated.csv')
seat=pd.read_csv('seat_updated.csv')
show=pd.read_csv('../show.csv')
```

```python
print(screen.shape)
screen.sample()
```

```python
print(seat.shape)
seat.sample()
```

```python
print(show.shape)
show.sample()
```

```python
screen_seat = screen.merge(seat, how='outer', on=['영화명', '점유율집계일', '누적관객수', '누적매출액'])
print(screen_seat.shape)
screen_seat.sample()
```

```python
index = []
for i, x in (screen_seat[['관객수_x', '관객수_y']]).iterrows():
    if x[0] != x[1]:
#         print(i, x)
        index.append(i)
    
screen_seat.iloc[index]
```

```python
index2 = []
for i, x in (screen_seat[['관객수_x', '관객수_y']]).iterrows():
    if x[0] != x[1]:
#         print(i, x)
        index2.append(i)
```

```python
screen_seat.iloc[index2]
```

```python
screen_seat = screen_seat.drop(['개봉일_y', '매출액_y', '관객수_y'], axis=1
                              ).rename(columns={'순위_x':'스크린점유율순위'
                            , '순위_y':'좌석판매점유율순위'
                            , '매출액_x':'매출액'
                            , '관객수_x':'관객수'
                            , '개봉일_x':'개봉일'
                           })
```

```python
screen_seat_show = screen_seat.merge(show, how='outer', on=['영화명', '점유율집계일', '누적관객수', '누적매출액'])

print(screen_seat_show.shape)
screen_seat_show.sample()
```

```python
index = []
for i, x in (screen_seat_show[['매출액_x', '매출액_y']]).iterrows():
    if x[0] != x[1]:
#         print(i, x)
        index.append(i)
    
screen_seat.iloc[index]
```

# Data

```python
audits = pd.read_csv('Data/audits_2015_ko.csv')
show_time = pd.read_csv('Data/showTm_2015_ko.csv')
companys = pd.read_csv('Data/companys_2015_ko.csv')

actors = pd.read_csv('Data/actors_2015_ko.csv')
actor_code = pd.read_csv('Data/actorCd_2015_list.csv')
actor_detail = pd.read_csv('Data/actor_detail_2015_ko.csv')

directors = pd.read_csv('Data/directors_2015_ko.csv')
director_code = pd.read_csv('Data/directorCd_2015_list.csv')
director_detail = pd.read_csv('Data/director_detail_2015_ko.csv')

staffs = pd.read_csv('Data/staffs_2015_ko.csv')
writer_code = pd.read_csv('Data/writerCd_2015_list.csv')
writer_detail = pd.read_csv('Data/writer_detail_2015_ko.csv')

print('audits', audits.shape)
print('show_time', show_time.shape)
print('companys', companys.shape)

print('actors', actors.shape)
print('actor_code', actor_code.shape)
print('actor_detail', actor_detail.shape)

print('directors', directors.shape)
print('director_code', director_code.shape)
print('director_detail', director_detail.shape)

print('staffs', staffs.shape)
print('writer_code', writer_code.shape)
print('writer_detail', writer_detail.shape)
```

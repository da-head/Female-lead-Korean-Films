```python
audits_2015 = pd.read_csv('../Data/audits_2015_ko.csv')
show_time_2015 = pd.read_csv('../Data/showTm_2015_ko.csv')
companys_2015 = pd.read_csv('../Data/companys_2015_ko.csv')
boxoffice_2015 = pd.read_csv('../Data/KOBIS_boxoffice_2015_filtered.csv')

audits_2016 = pd.read_csv('../Data/audits_2016_ko.csv')
show_time_2016 = pd.read_csv('../Data/showTm_2016_ko.csv')
companys_2016 = pd.read_csv('../Data/companys_2016_ko.csv')
boxoffice_2016 = pd.read_csv('../Data/KOBIS_boxoffice_2016_filtered.csv')

audits_2017 = pd.read_csv('../Data/audits_2017_ko.csv')
show_time_2017 = pd.read_csv('../Data/showTm_2017_ko.csv')
companys_2017 = pd.read_csv('../Data/companys_2017_ko.csv')
boxoffice_2017 = pd.read_csv('../Data/KOBIS_boxoffice_2017_filtered.csv')

audits_2018 = pd.read_csv('../Data/audits_2018_ko.csv')
show_time_2018 = pd.read_csv('../Data/showTm_2018_ko.csv')
companys_2018 = pd.read_csv('../Data/companys_2018_ko.csv')
boxoffice_2018 = pd.read_csv('../Data/KOBIS_boxoffice_2018_filtered.csv')

audits_2019 = pd.read_csv('../Data/audits_2019_ko.csv')
show_time_2019 = pd.read_csv('../Data/showTm_2019_ko.csv')
companys_2019 = pd.read_csv('../Data/companys_2019_ko.csv')
boxoffice_2019 = pd.read_csv('../Data/KOBIS_boxoffice_2019_filtered.csv')

print('audits_2015', audits_2015.shape)
print('audits_2016', audits_2016.shape)
print('audits_2017', audits_2017.shape)
print('audits_2018', audits_2018.shape)
print('audits_2019', audits_2019.shape)

print('show_time_2015', show_time_2015.shape)
print('show_time_2016', show_time_2016.shape)
print('show_time_2017', show_time_2017.shape)
print('show_time_2018', show_time_2018.shape)
print('show_time_2019', show_time_2019.shape)

print('companys_2015', companys_2015.shape)
print('companys_2016', companys_2016.shape)
print('companys_2017', companys_2017.shape)
print('companys_2018', companys_2018.shape)
print('companys_2019', companys_2019.shape)

print('boxoffice_2015', companys_2015.shape)
print('boxoffice_2016', companys_2016.shape)
print('boxoffice_2017', companys_2017.shape)
print('boxoffice_2018', companys_2018.shape)
print('boxoffice_2019', companys_2019.shape)
```

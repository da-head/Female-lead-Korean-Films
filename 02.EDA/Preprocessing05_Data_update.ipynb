## F-rated 다시 구한 후 저장
성별 부분 안 건드렸는데도 변화가 있는 게 이상하지만..

```python
merged3['F-rated'].value_counts()
```

```python
f_rated = []
for i, row in merged3.iterrows():
    answer = 0
    if row['배우성별'] == '여자' : answer += 1
    if row['작가성별'] == '여자' : answer += 1
    if row['감독성별'] == '여자' : answer += 1
        
    f_rated.append(answer)
merged3['F-rated'] = f_rated
```

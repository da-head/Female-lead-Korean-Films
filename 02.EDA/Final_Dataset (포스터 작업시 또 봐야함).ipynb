## F등급 다시 구하기: 감독 중 1명이라도 여성이면 F count +1

```python
film['F-rated'] = 0

for i in range(len(film)):
    if (film.loc[i, '감독1성별'] == '여자') | (film.loc[i, '감독2성별'] == '여자') | (film.loc[i, '감독3성별'] == '여자') | (film.loc[i, '감독4성별'] == '여자') : film.loc[i, 'F-rated'] += 1
    if film.loc[i, '작가성별'] == '여자' : film.loc[i, 'F-rated'] += 1
    if film.loc[i, '배우성별'] == '여자' : film.loc[i, 'F-rated'] += 1
```

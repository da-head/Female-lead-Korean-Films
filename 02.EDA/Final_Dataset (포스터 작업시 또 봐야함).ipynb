# F등급 다시 구하기 (코드 수정)

```python
film['F-rated'] = 0

for i in range(len(film)):
    if film.loc[i, '감독성별'] == '여자' : film.loc[i, 'F-rated'] += 1
    if film.loc[i, '작가성별'] == '여자' : film.loc[i, 'F-rated'] += 1
    if film.loc[i, '배우성별'] == '여자' : film.loc[i, 'F-rated'] += 1
```

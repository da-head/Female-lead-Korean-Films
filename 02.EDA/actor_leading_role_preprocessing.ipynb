### leading role 찾기

```python
# 각 영화코드를 딱 하나씩만 리스트에 넣는다.
moviecode = list(actors['movieCd'].unique())
 
# moviecode에서 원소를 하나씩 빼고나면 actors 행과의 비교가 끝나기 전에 리스트가 비어서 'list index out of range' 에러가 나므로 마지막 원소로 0을 추가해 놓는다.
moviecode.append(0) 

leading_role = actors.copy()

for i in range(len(leading_role)):
    
    # 동일한 영화코드를 가진 행들 중 첫번째 데이터를 보존하고, moviecode 리스트에서 그 코드를 제거한다.
    if leading_role.loc[i, 'movieCd'] == moviecode[0]:
        moviecode.pop(0)
    
    # 두번째 데이터부터 데이터프레임에서 삭제한다.
    else:
        leading_role = leading_role.drop(pd.DataFrame(leading_role.loc[i]), axis=0)
```

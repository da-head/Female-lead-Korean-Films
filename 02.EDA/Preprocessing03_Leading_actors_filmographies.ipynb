
# 주연배우 필모그래피 컬럼 merge

```python
# 배우 필모그래피 컬럼이 있는 파일 불러오기
actor = pd.read_csv('../01.Data_Collection/actor_code.csv')
actor.head()
```

```python
actor.isnull().sum()
```

```python
len(actor)
```

```python
# movieNm, peopleNm, repRoleNm 컬럼 삭제한 후 중복값 제거
actor = actor.drop(['movieNm', 'peopleNm', 'repRoleNm'], axis=1)
actor = actor.drop_duplicates()
len(actor)
```

```python
actor['peopleCd'].nunique()
```

```python
# 한 배우의 필모그래피가 여러 번 들어간 경우가 있다.
num_cd = actor.groupby('peopleCd')['filmoNames'].count().reset_index()
num_cd = num_cd[num_cd['filmoNames'] > 1].sort_values('peopleCd')
num_cd
```

```python
# 여러 행을 가진 배우코드만 따로 떼어서 보기
# 필모그래피에 들어가는 영화의 개수나 순서에 약간의 차이가 있어 중복값 제거 시 지워지지 않고 개별 행으로 남았음
dupe = actor[actor['peopleCd'].isin(num_cd['peopleCd'])].sort_values('peopleCd')
dupe
```

```python
# 배우코드별로 필모그래피에 포함된 영화 수가 가장 많은 행의 인덱스를 저장한다.

inclu_idx = []

for cd in dupe['peopleCd'].unique():
    filmo = dupe[dupe['peopleCd']==cd]
    inclu_idx.append(filmo['filmoNames'].apply(lambda x: len(x.split('|'))).idxmax())
```

```python
len(inclu_idx)
```

```python
# 나머지 행은 actor 테이블에서 제거한다.
exclu_idx = set(dupe.index) - set(inclu_idx)
actor = actor.drop(exclu_idx, axis=0)
```

```python
# 이제 배우코드별로 데이터 하나씩만 들어가 있는 것 확인
print(len(actor))
print(actor['peopleCd'].nunique())
```

```python
actor.sort_values('peopleCd').head()
```

```python
# df와 merge
df = df.merge(actor[['peopleCd', 'filmoNames']], how='left', left_on='배우코드', right_on='peopleCd')
```

```python
df.shape
```

```python
# drop 'peopleCd'
# 'filmoNames' 컬럼명을 '배우필모'로 변경
df = df.drop('peopleCd', axis=1).rename(columns = {'filmoNames': '배우필모'})
```

```python
f_rated = []
for i, row in df.iterrows():
    answer = 0
    if row['배우성별'] == '여자' : answer += 1
    if row['작가성별'] == '여자' : answer += 1
    if row['감독성별'] == '여자' : answer += 1
        
    f_rated.append(answer)
df['F-rated'] = f_rated
```

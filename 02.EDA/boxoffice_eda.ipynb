# 영화 EDA

```python
boxoffice.dtypes
```

```python
# 공백과 ',' 때문에 형변환이 안돼서 제거 작업.
boxoffice['sales']=boxoffice['sales'].str.replace(',', '')
boxoffice['Acc sales']=boxoffice['Acc sales'].str.replace(',', '')
boxoffice['Nu viewers']=boxoffice['Nu viewers'].str.replace(',', '')
boxoffice['Nu show']=boxoffice['Nu show'].str.replace(',', '')
boxoffice['Nu of screens']=boxoffice['Nu of screens'].str.replace(',', '')
boxoffice['production company']=boxoffice['production company'].str.replace(' ', '')
boxoffice['distributor']=boxoffice['distributor'].str.replace(' ', '')
```

```python
boxoffice['Acc viewers']=boxoffice['Acc viewers'].str.replace(',', '')
boxoffice['sales share']=boxoffice['sales share'].str.replace('%', '')
```

```python
boxoffice.head()
```

```python
boxoffice['sales'] = boxoffice['sales'].apply(lambda x: int(x))
boxoffice['Acc sales'] = boxoffice['Acc sales'].apply(lambda x: int(x))
boxoffice['Nu viewers'] = boxoffice['Nu viewers'].apply(lambda x: int(x))
boxoffice['Acc viewers'] = boxoffice['Acc viewers'].apply(lambda x: float(x))
boxoffice['Nu show'] = boxoffice['Nu show'].apply(lambda x: int(x))
boxoffice['production company'] = boxoffice['production company'].apply(lambda x: str(x))
boxoffice['distributor'] = boxoffice['distributor'].apply(lambda x: str(x))
```

```python
boxoffice.dtypes
```

```python
film = pd.read_csv('../01.Data_Collection/films_final.csv')
screen = pd.read_csv('../01.Data_Collection/screens_final.csv')
show = pd.read_csv('../01.Data_Collection/show_timetable.csv')
```

```python
film['장르1'] = film['장르'].apply(lambda x: x.split(',')[0])
```

```python
film['장르1'].unique()
```

```python
film[film['장르1'].isin(['공연', '뮤지컬', '기타'])]
```

```python
genre = film[~(film['장르1'].isin(['공연', '뮤지컬', '기타', '애니메이션', '다큐멘터리']))]
len(genre)
```

```python
genre['개봉일'].dtype
```

```python
genre['개봉일'] = pd.to_datetime(genre['개봉일'])
```

```python
y2015_1 = genre[(genre['개봉일'].dt.year==2015) & (genre['개봉일'].dt.month<=6)]
len(y2015_1)
```

```python
y2015_2 = genre[(genre['개봉일'].dt.year==2015) & (genre['개봉일'].dt.month>=7)]
y2016_1 = genre[(genre['개봉일'].dt.year==2016) & (genre['개봉일'].dt.month<=6)]
y2016_2 = genre[(genre['개봉일'].dt.year==2016) & (genre['개봉일'].dt.month>=7)]
y2017_1 = genre[(genre['개봉일'].dt.year==2017) & (genre['개봉일'].dt.month<=6)]
y2017_2 = genre[(genre['개봉일'].dt.year==2017) & (genre['개봉일'].dt.month>=7)]
y2018_1 = genre[(genre['개봉일'].dt.year==2018) & (genre['개봉일'].dt.month<=6)]
y2018_2 = genre[(genre['개봉일'].dt.year==2018) & (genre['개봉일'].dt.month>=7)]
y2019_1 = genre[(genre['개봉일'].dt.year==2019) & (genre['개봉일'].dt.month<=6)]
y2019_2 = genre[(genre['개봉일'].dt.year==2019) & (genre['개봉일'].dt.month>=7)]
```

```python
y15_1 = y2015_1.groupby('F-rated')['전국관객수'].sum().reset_index()
y15_1
```

```python
y15_2 = y2015_2.groupby('F-rated')['전국관객수'].sum().reset_index()
y16_1 = y2016_1.groupby('F-rated')['전국관객수'].sum().reset_index()
y16_2 = y2016_2.groupby('F-rated')['전국관객수'].sum().reset_index()
y17_1 = y2017_1.groupby('F-rated')['전국관객수'].sum().reset_index()
y17_2 = y2017_2.groupby('F-rated')['전국관객수'].sum().reset_index()
y18_1 = y2018_1.groupby('F-rated')['전국관객수'].sum().reset_index()
y18_2 = y2018_2.groupby('F-rated')['전국관객수'].sum().reset_index()
y19_1 = y2019_1.groupby('F-rated')['전국관객수'].sum().reset_index()
y19_2 = y2019_2.groupby('F-rated')['전국관객수'].sum().reset_index()
```

```python
y16_1 = y2016_1['F-rated'].value_counts().reset_index()
```

```python
y16_2 = y2016_2['F-rated'].value_counts().reset_index()
```

```python
y17_1 = y2017_1['F-rated'].value_counts().reset_index()
```

```python
y17_2 = y2017_2['F-rated'].value_counts().reset_index()
```

```python
y18_1 = y2018_1['F-rated'].value_counts().reset_index()
```

```python
y18_2 = y2018_2['F-rated'].value_counts().reset_index()
```

```python
y19_1 = y2019_1['F-rated'].value_counts().reset_index()
```

```python
y19_2 = y2019_2['F-rated'].value_counts()
y19_2
```

```python
count_f = y15_1.merge(y15_2, how='right', on='F-rated').merge(y16_1, how='left', on='F-rated').merge(y16_2, how='left', on='F-rated').merge(y17_1, how='left', on='F-rated').merge(y17_2, how='left', on='F-rated').merge(y18_1, how='left', on='F-rated').merge(y18_2, how='left', on='F-rated').merge(y19_1, how='left', on='F-rated').merge(y19_2, how='left', on='F-rated')
```

```python
count_f.loc[4] = count_f.loc[0] + count_f.loc[1] + count_f.loc[2] + count_f.loc[3]
```

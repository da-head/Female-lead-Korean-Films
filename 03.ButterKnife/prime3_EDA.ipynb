```python
weekly_audi = []
for movie in rate['영화명'].unique():

    sum_audi = 0
    for i, row in pd.DataFrame(rate.loc[rate['영화명'] == movie, ['개봉n일차(int)', '관객수']]).iterrows():
    #     print(row['개봉n일차(int)'], row['관객수'])

        if row['개봉n일차(int)'] % 7 == 0:
            n = row['개봉n일차(int)'] // 7
            weekly_audi.append((movie, n + 1, sum_audi))
            sum_audi = 0

        sum_audi += row['관객수']
```

```python
data = pd.DataFrame(weekly_audi, columns=['영화명', '개봉n주차', '주차관객수'])
print(data.shape)
data.sample()
```

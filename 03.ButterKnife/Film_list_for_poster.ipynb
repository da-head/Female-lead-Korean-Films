```python
title['영화코드'] = title['영화코드'].astype(str)
```

```python
writer = writer[writer['moviePartNm']=='시나리오(각본)']
```

```python
title = title.merge(writer, how='left', left_on='영화코드', right_on='movieCd')
```

```python
title = title.drop_duplicates().sort_values('개봉일', ascending=False)
```

```python
# 한 영화의 최대 감독 수: 4명
# 한 영화의 최대 각본가 수: 5명
title.groupby('영화코드')['peopleCd'].count().values.max()
```

```python
count = title[['영화코드', 'peopleCd', 'sex']]
```

```python
count_pivot = count.pivot_table(index='영화코드', columns='sex', aggfunc='count').rename(columns={'남자':'남각본', '여자':'여각본'})
count_pivot.head()
```

```python
title = title.merge(count_pivot, how='left', on='영화코드')
```
